```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster


```

```python
# Step 2: Load the data
data = pd.read_csv('Mall_Customers.csv')
data
```

```python
# Step 3: Data Preprocessing
# Check for missing values
print("Missing values in each column:")
print(data.isnull().sum())
```

```python
# Check for duplicates
print(f"Number of duplicate rows: {data.duplicated().sum()}")
data.drop_duplicates(inplace=True)
```

```python
# Data Cleaning: Drop unnecessary columns
data_cleaned = data.drop(columns=['CustomerID'])
```

```python
X = data_cleaned[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]  # Define the features you want to scale
```

```python
# Encode categorical variables
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
```

```python
#Check for NaN or Infinite Values
# Select only numeric columns
numeric_data = data_cleaned.select_dtypes(include=[np.number])

# Check for infinite values in numeric columns
print(np.isinf(numeric_data).sum())  # Count of infinite values per column
```

```python
#Handle Infinite Values: Replace or remove
data_cleaned.replace([np.inf, -np.inf], 0, inplace=True)
```

```python
# Step 4: Handling Outliers
# Boxplot for initial outlier detection
# Define features for the boxplot
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
```

```python
# Identify outliers based on IQR
Q1 = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].quantile(0.25)
Q3 = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].quantile(0.75)
IQR = Q3 - Q1

outliers = data[((data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] < (Q1 - 1.5 * IQR)) |
                  (data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))

# Create a boxplot for each feature
sns.boxplot(data=data[features], palette='Set2')

# Highlight outliers
for feature in features:
    outlier_data = outliers[feature]
    plt.scatter([features.index(feature)] * len(outlier_data),  # X positions for each feature
                outlier_data,
                color='red', marker='o', s=100, label='Outliers' if feature == features[0] else "")  # Avoid duplicate legend entries

# Set titles and labels
plt.title('Boxplot with Outliers Highlighted')
plt.xticks(ticks=range(len(features)), labels=features)
plt.legend()
plt.grid()

# Show the plot
plt.show()

```

```python
# Remove outliers based on IQR
Q1 = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].quantile(0.25)
Q3 = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].quantile(0.75)
IQR = Q3 - Q1
data_no_outliers = data[~((data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] < (Q1 - 1.5 * IQR)) |
                          (data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Boxplot for outlier detection after removal
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_no_outliers[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
plt.title('Boxplot for Outlier Detection (After Removal)')
plt.show()
```

```python
# Step 5: Exploratory Data Analysis (EDA)
# Univariate Analysis
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=15, kde=True)
plt.title('Distribution of Age')
plt.grid()
plt.show()
```

```python
# Bivariate Analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', palette='Set1')
plt.title('Annual Income vs. Spending Score')
plt.grid()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='crest', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Step 6: Data Scaling
scaler = MinMaxScaler()
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
scaled_features = scaler.fit_transform(features)
```

```python
# Step 7: Clustering
# Hierarchical Clustering
# Linkage matrix using 'ward' method
linkage_matrix = linkage(scaled_features, method='ward')
```

```python
# Plot Dendrogram
plt.figure(figsize=(10, 7))
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Customers")
plt.ylabel("Euclidean distances")
dendrogram(linkage_matrix)
plt.show()
```

```python
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_scores.append(score)
```

```python
# Plotting silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score for Different Number of Clusters (K-Means)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid()
plt.show()
```

```python
# Using optimal k from the elbow method
optimal_k = 5  # Replace with your chosen k based on the Elbow Method
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['KMeans_Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Clustering using DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
data['DBSCAN_Cluster'] = dbscan.fit_predict(scaled_features)
```

```python
# Clustering using Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=optimal_k)
data['Agglo_Cluster'] = agglo.fit_predict(scaled_features)
```

```python
# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust the number of clusters as needed
kmeans_labels = kmeans.fit_predict(scaled_features)
```

```python
# Step 8: Performance Evaluation
# Evaluate K-Means clustering performance
kmeans_silhouette = silhouette_score(scaled_features, data['KMeans_Cluster'])
kmeans_db_index = davies_bouldin_score(scaled_features, data['KMeans_Cluster'])
kmeans_ch_index = calinski_harabasz_score(scaled_features, data['KMeans_Cluster'])
```

```python
print("\nK-Means Clustering Performance:")
print(f"Silhouette Score: {kmeans_silhouette:.4f}")
print(f"Davies-Bouldin Index: {kmeans_db_index:.4f}")
print(f"Calinski-Harabasz Index: {kmeans_ch_index:.4f}")
```

```python
# Predict on test set
y_pred = rf.predict(X_test)
```

```python
# Accuracy of Random Forest
rf_accuracy = accuracy_score(y_test, y_pred)
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")
```
